```python
# Question 2.
# NOTE: You need to add alldrugprops.json to the file directory to continue.
import json
import pandas as pd

with open('alldrugsprops.json') as f:
  drugProps = json.load(f)

data = []
for prop in drugProps[0:10]:
  data.append(prop)

df = pd.DataFrame(data=data)
print(df.to_string())
```

```python
logp, psa, qed = [], [], [] # initialize lists
for prop in drugProps: # iterate through each item in drugProps
  logp.append(prop["LOGP"])
  psa.append(prop["PSA"])
  qed.append(prop["QED"])


fig, ax = plt.subplots()
ax.scatter(
    x = logp,
    y = psa,
    c = qed
)

ax.set_xlabel("logp")
ax.set_ylabel("psa")
plt.show()
```
